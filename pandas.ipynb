**CSV**

```python
from io import StringIO,BytesIO
```

```python
data = ('col1,col2,col3\n'
        'x,y,1\n'
        'a,b,2\n'
        '4,5,6')
```

```python
data
```

```python
pd.read_csv(StringIO(data))
```

```python
# read from specific column
df= pd.read_csv(StringIO(data),usecols=['col2','col3'])
```

```python
df
```

```python
#convert dataframe to csv
df.to_csv('test.csv')
```
